## Exercício 1:
Utilizando a base de dados de covid:

**a)** Faça um gráfico de barras mostrando o total de óbitos (considerando todos os meses) para todos os DRS, excluindo a Grande São Paulo. Coloque título no gráfico e nos eixos, defina um tamanho de imagem bom e intercale as barras com as cores 'red' e 'green'.

```python
covid_exceto_sp = df2[df2['nome_drs']!="Grande São Paulo"].groupby(['nome_drs'], as_index = False).sum()
covid_exceto_sp.plot(x = 'nome_drs', y ='obitos_novos', kind='bar', title = 'Novas mortes por covid',
                     label='Óbitos', xlabel ='Nome do DRS', ylabel = 'Total de óbitos', color=['red', 'green'], figsize=(12,6))
```

**b)**  Faça um gráfico de barras horizontal mostrando o total de óbitos no mês de abril para 3 DRS da sua escolha. Coloque título no gráfico e nos eixos, defina um tamanho de imagem bom.

```python
DRS_escolhidos = ['Baixada Santista', 'Franca', 'Sorocaba']

df2[df2['nome_drs'].isin(DRS_escolhidos)
    & (df2.ano_mes =='2020-04')].groupby(['nome_drs'], as_index =False).sum().plot(x='nome_drs',
                                                                                   y='obitos_novos',kind='barh',
                                                                                   figsize=(10,5),
                                                                                   title='Óbitos por DRS no mês de abril/20',
                                                                                   label='óbitos',
                                                                                   xlabel='Nome DRS')
```
